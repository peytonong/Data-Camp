### 4. Effectiveness of Marketing (Time used for each sales for each of the methods)

From the project description, we knew that:
- **Email**: Required very little work for the team. I assumed it took **_1 minute_** for sales approached via email while the time of 1 minute is certainly overestimated as email blasting only took seconds but the assumption is to align a same number format and decimal for time taken for all three methods
- **Call**: Customers were called by a member of sales team. The average time needed for each customers is **_30 minutes_**
- **Email and Call**: Customers were first sent an email with the product information first then called a week later by a member of sales team. The average time needed for each customers is **_10 minutes_**

With all the assumptions, an overview on effectiveness of each marketing method has been conducted.

Result:
- Email has the highest mean of number of sold per minute and highest mean of revenue per minute suggesting that email marketing has the highest effectiveness, following by "Email and Call" together and lastly "Call" method.

```python
time_needed = {
    'Call': 30,          # in minutes
    'Email': 1,          # in minutes
    'Email + Call': 10   # in minutes
}

data['time_needed'] = data['sales_method'].map(time_needed)

data['nb_sold_per_minute'] = data['nb_sold'] / data['time_needed']
data['revenue_per_minute'] = data['revenue'] / data['time_needed']

sales_effectiveness = data.groupby('sales_method').agg(
    mean_nb_sold_per_minute=('nb_sold_per_minute', 'mean'),
    mean_revenue_per_minute=('revenue_per_minute', 'mean')
).reset_index()

sales_effectiveness = sales_effectiveness.sort_values(by='mean_revenue_per_minute', ascending=False)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.35  # the width of the bars

# Position of the bars on the x-axis
r1 = range(len(sales_effectiveness))
r2 = [x + width for x in r1]

# Create bars for mean_nb_sold_per_minute
ax.bar(r1, sales_effectiveness['mean_nb_sold_per_minute'], color='blue', width=width, edgecolor='grey', label='Mean Nb Sold Per Minute')

# Create bars for mean_revenue_per_minute
ax.bar(r2, sales_effectiveness['mean_revenue_per_minute'], color='green', width=width, edgecolor='grey', label='Mean Revenue Per Minute')

# Add labels
ax.set_xlabel('Sales Method', fontweight='bold')
ax.set_title('Sales Effectiveness by Method (Per Minute)')
ax.set_xticks([r + width / 2 for r in range(len(sales_effectiveness))])
ax.set_xticklabels(sales_effectiveness['sales_method'])

# Create legend & Show graphic
ax.legend()
plt.show()

print(sales_effectiveness)
```

 ### 5. Which states have higher revenues?
 
Total revenue per state is being summarized and as the result below, it is found that the top 3 states with highest revenue are California, Texas, and New York. 

```python
import folium
import geopandas as gpd
import pandas as pd
from folium import Choropleth, LayerControl

state_revenues = data.groupby('state').agg(
    sum_revenue = ('revenue','sum'),
    mean_revenue = ('revenue','mean'),
    median_revenue = ('revenue','median')
).reset_index()
print(state_revenues.sort_values(by="sum_revenue",ascending=False).head(5))

# Load US states geometries
url = 'https://raw.githubusercontent.com/PublicaMundi/MappingAPI/master/data/geojson/us-states.json'
geo_data = gpd.read_file(url)

# Merge data with the GeoDataFrame
geo_data['name'] = geo_data['name'].str.title()  
# Ensure state names match the format in your data
merged = geo_data.set_index('name').join(state_revenues.set_index('state'), how='inner')

# Initialize Folium map
m = folium.Map(location=[37.8, -96.9], zoom_start=4)

# Add Choropleth layer
choropleth = folium.Choropleth(
    geo_data=geo_data,
    name='Choropleth',
    data=merged,
    columns=[merged.index, 'sum_revenue'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',  # Using a valid ColorBrewer scheme
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Total Revenue',
    nan_fill_color='white'
).add_to(m)

# Add labels
folium.GeoJson(
    geo_data,
    style_function=lambda feature: {
        'fillColor': 'transparent',
        'color': 'transparent'
    },
    tooltip=folium.GeoJsonTooltip(
        fields=['name'],
        aliases=['State: '],
        localize=True
    )
).add_to(m)

# Add Layer Control
folium.LayerControl().add_to(m)

# Save the map as an HTML file
m.save('total_revenue_map.html')

# Show the map (in a Jupyter notebook, you can simply display it)
m
```

### 7. Is there any relationship between years as customers and the median revenue achieved by each sales method?

The data is grouped into three categories based on the years as customer, 0-5 years, 5-10 years, and more than 10 years. The median revenues achieved by each sales method are calculated for each of these categories.

From the data, we can see that there is not a significant difference in the median revenues achieved by each sales method across the three categories. The revenues achieved by each sales method are relatively consistent across the three categories, with only slight variations.

Therefore, we can conclude that there is not a strong relationship between customer years with the company and the median revenues achieved by each sales method. There is only a small increase in revenue by newer customers group compared to others.

```python
data['years_range'] = pd.cut(data['years_as_customer'], bins=[0, 5, 10, 100], 
                             labels=['0-5', '5-10', '10+'])

med_revenue_by_years_sales = data.groupby(['years_range', 'sales_method'])['revenue'].median().reset_index()

line_chart_data = med_revenue_by_years_sales.pivot(index='years_range', columns='sales_method', values='revenue')

line_chart_data.plot(kind='line', marker='o', color=colors1)
plt.title('Median Revenue by Years Range and Sales Method')
plt.xlabel('Years Range')
plt.ylabel('Median Revenue')
plt.show()

print(med_revenue_by_years_sales.head())
```

### 8. Is there any relationship between years as customers and median units sold by each sales method?

### 9. Is there any relationship between number of site visit and revenue achieved by each sales method over time?

Based on the updated data provided, it appears that there is a positive relationship between the number of site visits and the revenue achieved by each sales method over time. 

Analyzing the data, we can see that the revenue achieved by the 'Email + Call' method is the highest among the three sales methods for each week. 

Furthermore, we can observe that as the number of site visits increases, the revenue also increases for each week and sales method.

```python
grouped_data = data.groupby(['week', 'sales_method']).agg({'revenue': 'median', 'nb_site_visits': 'median'}).reset_index()

colors = ["mediumslateblue","lightsteelblue","deepskyblue"]

plt.scatter(x=grouped_data.loc[grouped_data['sales_method'] == 'Call', 'week'], 
            y=grouped_data.loc[grouped_data['sales_method'] == 'Call', 'revenue'], 
            s=grouped_data.loc[grouped_data['sales_method'] == 'Call', 'nb_site_visits']*10,
            c=colors[0],
            alpha=0.5,
            label='Call')

plt.scatter(x=grouped_data.loc[grouped_data['sales_method'] == 'Email', 'week'], 
            y=grouped_data.loc[grouped_data['sales_method'] == 'Email', 'revenue'], 
            s=grouped_data.loc[grouped_data['sales_method'] == 'Email', 'nb_site_visits']*10,
            c=colors[1],
            alpha=0.5,
            label='Email')

plt.scatter(x=grouped_data.loc[grouped_data['sales_method'] == 'Email + Call', 'week'], 
            y=grouped_data.loc[grouped_data['sales_method'] == 'Email + Call', 'revenue'], 
            s=grouped_data.loc[grouped_data['sales_method'] == 'Email + Call', 'nb_site_visits']*10,
            c=colors[2],
            alpha=0.5,
            label='Email + Call')

plt.title('Relationship Between Number of Site Visits and Revenue by Sales Method Over Time')
plt.xlabel('week')
plt.ylabel('Revenue')
plt.legend(title='Sales Method')

plt.show()
print(grouped_data.head())
```

## Business Metric

To enhance sales performance and resource allocation, it is essential to determine the most effective sales method. Given the varying time commitments associated with different approaches, it is crucial to assess their performance in relation to the resources invested.

Key metrics to consider include:

1. Customer Acquisition Cost:
- Measures the efficiency of acquiring new customers.
- While "Email" was found to be the most time-efficient method in previous analysis, the absence of cost data prevents a comprehensive evaluation of its cost-effectiveness.
- This metric was chosen because by minimizing the customer acquisition cost, the resources used for each product can be minimized, allowing the business to expand its product line and introduce new products constantly without concern over sales. Additionally, this metric is easily tracked as the required data consists only of time and cost needed for production and their respective sales, which are highly accessible and trackable, easing the continuity of metric tracking.

2. Customer Lifetime Value:
- Evaluates the long-term profitability of customer segments.
